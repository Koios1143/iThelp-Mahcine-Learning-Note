```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lr = 0.000001
epochs = 1000
x = np.array([498, 523, 550, 834, 980, 1250, 1430, 1700, 1780, 1900, 2100])
y = np.array([100, 150, 200, 210, 300, 290, 340, 293, 350, 290, 285])
n = len(x)
a, b = 0, 0
Losses = []
```

```python
def Loss(y, y_pred):
    return np.sum((y - y_pred) ** 2) / (2 * n)
```

```python
for i in range(epochs):
    y_pred = a*x + b
    loss = Loss(y, y_pred)
    da = np.sum(x * (y_pred - y)) / n
    db = np.sum(y_pred - y) / n
    a = a - lr * da
    b = b - lr * db
    print(f'epoch {i}\tLoss: {loss}\ta: {a}\tb: {b}')
```

```python
plt.figure(figsize=(10, 5))
plt.subplots_adjust(left=0.3, right=0.8)
plt.scatter(x, y)
plt.plot([x for x in range(498, 2101)], [a*x + b for x in range(498, 2101)])
plt.savefig('regression_result.png')
```

```python
plt.plot([i for i in range(epochs)], Losses)
plt.show()
```
