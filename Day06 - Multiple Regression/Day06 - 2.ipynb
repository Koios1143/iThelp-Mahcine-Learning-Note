```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd
```

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
x = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
y = raw_df.values[1::2, 2]
labels = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV']
n = len(y)
features = x.shape[1]
```

```python
lr = 0.00001
epochs = 1000
Losses = []
```

```python
W = np.zeros((features, 1))
c = 0
```

```python
def Loss(y, y_pred):
    return np.sum((y - y_pred) ** 2) / (2 * n)
```

```python
for i in range(epochs):
    y_pred = np.dot(x, W) + c
    loss = Loss(y, y_pred)
    Losses.append(loss)
    
    dW = np.array([np.sum(W[i] * (y_pred - y)) / n for i in range(len(W))]).reshape(len(W), 1)
    dc = np.sum(y_pred - y) / n
    
    W = W - lr * dW
    c = c - lr * dc
    print(f'epoch {i}\tLoss: {loss}')
```

```python
plt.figure(figsize=(10, 5))
plt.subplots_adjust(left=0.3, right=0.8)
plt.plot([i for i in range(epochs)], Losses)
plt.savefig('Loss2.png')
```
